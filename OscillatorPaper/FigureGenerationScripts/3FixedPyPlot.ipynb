```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import Normalize
import seaborn as sns
```

```python
# Read the CSV file
file_path = '/home/local/WIN/jfisch27/Desktop/Julia/OscillatorPaper/FigureGenerationScripts/fixed_triplet_results-kcat1kcat7DF.csv'
data = pd.read_csv(file_path)
```

```python
# Clean the data by removing extreme values
data = data.replace([np.inf, -np.inf], np.nan).dropna()
data = data[(data != 0).all(1)]
data = data[(np.log10(data) > -300).all(1)]
data
```

```python
# Log scale the parameters
log_kcat1_cleaned = np.log10(data['kcat1'])
log_kcat7_cleaned = np.log10(data['kcat7'])
log_DF_cleaned = np.log10(data['DF'])
```

```python
# Normalize the sizes for the scatter plot
sizes = Normalize()(data['average_amplitude'])

# Define a colormap for the plots
cmap = 'viridis'

# Define angles for different views
angles = [0, 30, 60, 90, 120, 150]
```

```python
# Set a colormap
cmap = plt.get_cmap('viridis')

# Reload the dataset
fixed_triplet_results_kcat1kcat7DF = data

# Define a threshold to identify extreme values (values close to zero)
threshold = 1e-8

# Replace extreme values with 0 in the relevant columns
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['kcat1'] < threshold, 'kcat1'] = 0
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['kcat7'] < threshold, 'kcat7'] = 0
fixed_triplet_results_kcat1kcat7DF.loc[fixed_triplet_results_kcat1kcat7DF['DF'] < threshold, 'DF'] = 0

# Define the log-scaled parameters (with epsilon added to avoid taking log of zero)
epsilon = 1e-8
log_kcat1_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['kcat1'] + epsilon)
log_kcat7_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['kcat7'] + epsilon)
log_DF_cleaned = np.log10(fixed_triplet_results_kcat1kcat7DF['DF'] + epsilon)

# Normalize the average amplitude and period for the point size and color
fixed_triplet_results_kcat1kcat7DF['average_amp_norm'] = (fixed_triplet_results_kcat1kcat7DF['average_amplitude'] - fixed_triplet_results_kcat1kcat7DF['average_amplitude'].min()) / (fixed_triplet_results_kcat1kcat7DF['average_amplitude'].max() - fixed_triplet_results_kcat1kcat7DF['average_amplitude'].min())
fixed_triplet_results_kcat1kcat7DF['average_per_norm'] = (fixed_triplet_results_kcat1kcat7DF['average_period'] - fixed_triplet_results_kcat1kcat7DF['average_period'].min()) / (fixed_triplet_results_kcat1kcat7DF['average_period'].max() - fixed_triplet_results_kcat1kcat7DF['average_period'].min())

# Create 3D scatter plot with projections onto each plane
fig = plt.figure(figsize=(14, 7))
ax = fig.add_subplot(111, projection='3d')

# Create the main 3D scatter plot
sc = ax.scatter(log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned,
                c=fixed_triplet_results_kcat1kcat7DF['average_per_norm'],
                s=fixed_triplet_results_kcat1kcat7DF['average_amp_norm']*500,
                alpha=0.6, cmap=cmap, marker='o')

# Project the points onto each plane and connect them with lines
for x, y, z in zip(log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned):
    ax.plot([x, x], [y, y], [z, log_DF_cleaned.min()], c='gray', alpha=0.2) # Projection onto XY plane
    ax.plot([x, x], [y, log_kcat7_cleaned.max()], [z, z], c='gray', alpha=0.2) # Projection onto XZ plane
    ax.plot([x, log_kcat1_cleaned.max()], [y, y], [z, z], c='gray', alpha=0.2) # Projection onto YZ plane

# Set larger axis labels
ax.set_xlabel('log10(kcat1)', fontsize=14)
ax.set_ylabel('log10(kcat7)', fontsize=14)
ax.set_zlabel('log10(DF)', fontsize=14)

# Add a colorbar
plt.colorbar(sc, ax=ax, label='Average Period')

# Set the title
ax.set_title('Log-Scaled Parameters with Projections and Lines (Cleaned Data) - Size: Average Amplitude, Color: Average Period', fontsize=14)

plt.show()
```

```python
# Function to create 3D scatter plot with accurate color-matching projections (shadows) on all walls
def create_3d_scatter_with_accurate_shadows(angle, x, y, z, sizes, colors):
    fig = plt.figure(figsize=(14, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Normalize the colors for accurate mapping
    norm_colors = Normalize()(colors)

    # Create the main 3D scatter plot with reduced alpha
    sc = ax.scatter(x, y, z, c=colors, s=sizes * 300, alpha=1.0, cmap=cmap, marker='o')

    # Get the colormap to extract RGBA values
    colormap = plt.cm.get_cmap(cmap)

    # Project the points onto each plane and connect them with lines, including color-matching shadows
    for xi, yi, zi, si, ci in zip(x, y, z, sizes, norm_colors):
        rgba_color = colormap(ci)
        ax.plot([xi, xi], [yi, y.min() -1], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(xi, y.min()-1, zi, color=rgba_color, s=si * 100, alpha=0.3, marker='x')
        ax.plot([xi, x.min()-1], [yi, yi], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(x.min()-1, yi, zi, color=rgba_color, s=si * 100, alpha=0.3, marker='x')
        ax.plot([xi, xi], [yi, yi], [z.min()-1, zi], c='gray', alpha=0.2)
        ax.scatter(xi, yi, z.min()-1, color=rgba_color, s=si * 100, alpha=0.3, marker='x')

    ax.set_xlabel('log10(kcat1)', fontsize=14)
    ax.set_ylabel('log10(kcat7)', fontsize=14)
    ax.set_zlabel('log10(DF)', fontsize=14)

    
    ax.view_init(elev=30, azim=angle)
    plt.colorbar(sc, ax=ax, label='Average Period')
    ax.grid(False)
    ax.set_title(f'Log-Scaled Parameters with Accurate Color-Matching Projections & Shadows (Angle: {angle}) - Size & Color: Average Amplitude & Period', fontsize=14)
    plt.show()
```

```python
create_3d_scatter_with_accurate_shadows(30, log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data.average_period)
```

```python
# Plot from different angles with log scaling and accurate color-matching shadows
for angle in angles:
    create_3d_scatter_with_accurate_shadows(angle, log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data['average_period'])
```

```python
# Create a function for 2D subplots
def create_2d_subplots(x, y, z, sizes, colors, angle):
    fig = plt.figure(figsize=(18, 12))
    
    # Main 3D scatter plot
    ax_main = fig.add_subplot(2, 2, 2, projection='3d')
    sc_main = ax_main.scatter(x, y, z, c=colors, s=sizes * 300, alpha=0.5, cmap=cmap, marker='o')
    ax_main.view_init(elev=30, azim=angle)
    ax_main.set_xlabel('log10(kcat1)', fontsize=14)
    ax_main.set_ylabel('log10(kcat7)', fontsize=14)
    ax_main.set_zlabel('log10(DF)', fontsize=14)
    plt.colorbar(sc_main, ax=ax_main, label='Average Period')
    ax_main.set_title('3D View', fontsize=16)
    
    # XY plane projection
    ax_xy = fig.add_subplot(2, 2, 1)
    sc_xy = ax_xy.scatter(x, y, c=colors, s=sizes * 300, alpha=0.5, cmap=cmap)
    ax_xy.set_xlabel('log10(kcat1)', fontsize=14)
    ax_xy.set_ylabel('log10(kcat7)', fontsize=14)
    ax_xy.set_title('Fixed kcat1 vs. kcat7', fontsize=16)
    
    # XZ plane projection
    ax_xz = fig.add_subplot(2, 2, 3)
    sc_xz = ax_xz.scatter(x, z, c=colors, s=sizes * 300, alpha=0.5, cmap=cmap)
    ax_xz.set_xlabel('log10(kcat1)', fontsize=14)
    ax_xz.set_ylabel('log10(DF)', fontsize=14)
    ax_xz.set_title('Fixed kcat1 vs. DF', fontsize=16)
    
    # YZ plane projection
    ax_yz = fig.add_subplot(2, 2, 4)
    sc_yz = ax_yz.scatter(y, z, c=colors, s=sizes * 300, alpha=0.5, cmap=cmap)
    ax_yz.set_xlabel('log10(kcat7)', fontsize=14)
    ax_yz.set_ylabel('log10(DF)', fontsize=14)
    ax_yz.set_title('Fixed kcat7 vs. DF', fontsize=16)
    
    plt.suptitle('3D Scatter Plot with Projections on XY, XZ, and YZ Planes', fontsize=18)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```

```python
# Plot for a selected angle
create_2d_subplots(log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data['average_period'], angles[0])
```

```python
# Function to plot the main 3D scatter plot with projections (shadows) on all walls
def plot_3d_scatter_with_projections(ax, x, y, z, sizes, colors, angle):
    # Normalize the colors for accurate mapping
    norm_colors = Normalize()(colors)

    # Create the main 3D scatter plot with reduced alpha
    sc = ax.scatter(x, y, z, c=colors, s=sizes * 100, alpha=0.8, cmap=cmap, marker='o')

    # Get the colormap to extract RGBA values
    colormap = plt.cm.get_cmap(cmap)

    plt.colorbar(sc, ax=ax, label='Average Period')


    # Project the points onto each plane and connect them with lines, including color-matching shadows
    for xi, yi, zi, si, ci in zip(x, y, z, sizes, norm_colors):
        rgba_color = colormap(ci)
        ax.plot([xi, xi], [yi, y.min() - 1], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(xi, y.min() - 1, zi, color=rgba_color, s=si * 100, alpha=0.3, marker='x')
        ax.plot([xi, x.min() - 1], [yi, yi], [zi, zi], c='gray', alpha=0.2)
        ax.scatter(x.min() - 1, yi, zi, color=rgba_color, s=si * 100, alpha=0.3, marker='x')
        ax.plot([xi, xi], [yi, yi], [z.min() - 1, zi], c='gray', alpha=0.2)
        ax.scatter(xi, yi, z.min() - 1, color=rgba_color, s=si * 100, alpha=0.3, marker='x')

    ax.set_xlabel('log10(kcat1)', fontsize=14)
    ax.set_ylabel('log10(kcat7)', fontsize=14)
    ax.set_zlabel('log10(DF)', fontsize=14)
    ax.view_init(elev=30, azim=angle)
    ax.grid(False)


# Create 3D scatter plot with accurate color-matching projections (shadows) on all walls and 2D subplots
def create_3d_scatter_with_accurate_shadows_and_subplots(angle, x, y, z, sizes, colors):
    fig = plt.figure(figsize=(20, 10))

    # Main 3D scatter plot
    ax = fig.add_subplot(2, 2, 1, projection='3d')
    plot_3d_scatter_with_projections(ax, x, y, z, sizes, colors, angle)
    

    # XY Plane
    ax_xy = fig.add_subplot(2, 2, 2)
    ax_xy.scatter(x, y, c=colors, s=sizes * 300, alpha=0.6, cmap=cmap)
    ax_xy.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xy.set_ylabel('log10(kcat7)', fontsize=12)
    ax_xy.set_title('kcat1 vs. kcat7', fontsize=14)

    # XZ Plane
    ax_xz = fig.add_subplot(2, 2, 3)
    ax_xz.scatter(x, z, c=colors, s=sizes * 300, alpha=0.6, cmap=cmap)
    ax_xz.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xz.set_ylabel('log10(DF)', fontsize=12)
    ax_xz.set_title('kcat1 vs. DF', fontsize=14)

    # YZ Plane
    ax_yz = fig.add_subplot(2, 2, 4)
    ax_yz.scatter(y, z, c=colors, s=sizes * 300, alpha=0.6, cmap=cmap)
    ax_yz.set_xlabel('log10(kcat7)', fontsize=12)
    ax_yz.set_ylabel('log10(DF)', fontsize=12)
    ax_yz.set_title('kcat7 vs. DF', fontsize=14)

    plt.tight_layout()
    plt.show()

# Plot for a selected angle
create_3d_scatter_with_accurate_shadows_and_subplots(angles[0], log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data['average_period'])
```

```python
# Create 3D scatter plot with accurate color-matching projections (shadows) on all walls and 2D subplots
def create_3d_scatter_with_violin_subplots(angle, x, y, z, sizes, colors):
    fig = plt.figure(figsize=(20, 10))

    # Main 3D scatter plot
    ax = fig.add_subplot(2, 2, 1, projection='3d')
    plot_3d_scatter_with_projections(ax, x, y, z, sizes, colors, angle)
    

    # XY Plane
    ax_xy = fig.add_subplot(2, 2, 2)
    ax_xy.violinplot([x, y])
    ax_xy.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xy.set_ylabel('log10(kcat7)', fontsize=12)
    ax_xy.set_title('XY Plane', fontsize=14)

    # XZ Plane
    ax_xz = fig.add_subplot(2, 2, 3)
    ax_xz.violinplot([x, z])
    ax_xz.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xz.set_ylabel('log10(DF)', fontsize=12)
    ax_xz.set_title('XZ Plane', fontsize=14)

    # YZ Plane
    ax_yz = fig.add_subplot(2, 2, 4)
    ax_yz.violinplot([y, z])
    ax_yz.set_xlabel('log10(kcat7)', fontsize=12)
    ax_yz.set_ylabel('log10(DF)', fontsize=12)
    ax_yz.set_title('YZ Plane', fontsize=14)

    plt.tight_layout()
    plt.show()

# Plot for a selected angle
create_3d_scatter_with_violin_subplots(angles[0], log_kcat1_cleaned, log_kcat7_cleaned, log_DF_cleaned, sizes, data['average_period'])
```

```python
def create_3d_scatter_with_histogram_subplots(angle, x, y, z, sizes, colors):
    fig = plt.figure(figsize=(20, 10))

    # Main 3D scatter plot
    ax = fig.add_subplot(2, 2, 1, projection='3d')
    plot_3d_scatter_with_projections(ax, x, y, z, sizes, colors, angle)

    # XY Plane
    ax_xy = fig.add_subplot(2, 2, 2)
    ax_xy.hist(x, bins=20, color = 'blue', edgecolor='black')
    ax_xy.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xy.set_ylabel('log10(kcat7)', fontsize=12)
    ax_xy.set_title('XY Plane', fontsize=14)

    # XZ Plane
    ax_xz = fig.add_subplot(2, 2, 3)
    ax_xz.hist(y, bins=20, color='green', edgecolor='black')
    ax_xz.set_xlabel('log10(kcat1)', fontsize=12)
    ax_xz.set_ylabel('log10(DF)', fontsize=12)
    ax_xz.set_title('XZ Plane', fontsize=14)

    # YZ Plane
    ax_yz = fig.add_subplot(2, 2, 4)
    ax_yz.hist(z, bins=20, color='red', edgecolor='black')
    ax_yz.set_xlabel('log10(kcat7)', fontsize=12)
    ax_yz.set_ylabel('log10(DF)', fontsize=12)
    ax_yz.set_title('YZ Plane', fontsize=14)

    plt.tight_layout()
    plt.show()
```
